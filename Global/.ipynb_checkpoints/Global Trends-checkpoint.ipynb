```python
Covid_Summary = pd.read_csv('https://raw.githubusercontent.com/datasets/covid-19/master/data/countries-aggregated.csv', parse_dates=['Date'])
Covid_Summary.head()
```

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go


import warnings
warnings.filterwarnings('ignore')
```

```python
Covid_Summary.isnull().sum()
```

```python
Covid_Summary.info()
```

```python
Covid_Summary['Active'] = Covid_Summary['Confirmed'] - Covid_Summary['Recovered'] - Covid_Summary['Deaths']
Covid_Summary.tail()
```

```python
Covid_Summary['Date'] = pd.to_datetime(Covid_Summary['Date'])
```

```python
Covid_Summary.head()
```

```python
Covid_Summary.info()
```

```python
cases = Covid_Summary[Covid_Summary['Date'] == Covid_Summary['Date'].max()].copy().fillna(0)
cases.index = cases["Country"]
cases = cases.drop(['Country','Date'], axis=1)
```

```python
cases.head()
```

```python
df = pd.DataFrame(pd.to_numeric(cases.sum())).transpose()
df.style.background_gradient(cmap='BuGn',axis=1)
```

```python
Trend = Covid_Summary.groupby(['Date'])['Confirmed', 'Deaths','Recovered','Active'].sum().reset_index()
```

```python
Trend.head()
```

```python
fig = go.Figure(go.Bar(x= Trend.Date, y= Trend.Recovered, name='Recovered Cases', marker_color='green'))
fig.add_trace(go.Bar(x=Trend.Date, y= Trend.Active, name='New Cases',marker_color='yellow'))
fig.add_trace(go.Bar(x=Trend.Date, y= Trend.Deaths, name='Deaths',marker_color='red'))

fig.update_layout(barmode='relative',legend_orientation="h",legend=dict(x= 0.3, y=1.1))
fig.show()
```

```python
%matplotlib inline
import plotly.express as px
```

```python
def horizontal_bar_chart(df, x, y, title, x_label, y_label, color):
    fig = px.bar(df, x=x, y=y, orientation='h', title=title, 
                 labels={x.name: x_label,
                         y.name: y_label}, color_discrete_sequence=[color])
    fig.update_layout(yaxis={'categoryorder': 'total ascending'})
    fig.show()
```

```python
top_10_death_states = cases.sort_values('Deaths',ascending = False)[:10]
horizontal_bar_chart(top_10_death_states,top_10_death_states.Deaths, top_10_death_states.index,
                     'Top 10 Countries with most deaths','Number of deaths(In Thousands)','State Name','Orange')
```

```python
top_10_confirmed_states = cases.sort_values('Confirmed', ascending=False)[:10]
horizontal_bar_chart(top_10_confirmed_states,top_10_confirmed_states.Confirmed, top_10_confirmed_states.index,
            'Top 10 Countries (Confirmed Cases)', 'Number of Confirmed cases (in Thousands)','States Name','blue')
```

```python
top_10_recovered_states = cases.sort_values('Recovered', ascending=False)[:10]
horizontal_bar_chart(top_10_recovered_states, top_10_recovered_states.Recovered, top_10_recovered_states.index,
                    'Top 10 Countries (Cured Cases)', 'Number of Cured cases (in Thousands)', 'States Name', 'Purple')
```

For the prediction purpose we have use Prophet library produced by Facebook which is used for Time series Forecasting. Prophet is a procedure for forecasting time series data based on an additive model where non-linear trends are fit with yearly, weekly, and daily seasonality, plus holiday effects. It works best with time series that have strong seasonal effects and several seasons of historical data.

```python
from fbprophet import Prophet
from fbprophet.plot import plot_plotly, add_changepoints_to_plot
from plotly.offline import iplot, init_notebook_mode
```

```python
model = Prophet(yearly_seasonality = False, weekly_seasonality = False, daily_seasonality = True)
```

```python
Confirmed = Trend.loc[:, ['Date', 'Confirmed']] 
Confirmed.tail()
```

```python
Cured = Trend.loc[:, ['Date', 'Recovered']] 
Cured.tail()
```

```python
# Dataframe must have columns "Date" and "Values" with the dates and values  for prophet prediction
Confirmed.columns = ['ds', 'y']
model.fit(Confirmed)
```

```python
future = model.make_future_dataframe(periods=60) # helper function to extend the dataframe for specified days
future.tail()
```

```python
# yhat represents the prediction, while yhat_lower and yhat_upper represent the lower and upper bound
forecast_conf = model.predict(future)
forecast_conf
```

```python
fig = plot_plotly(model, forecast_conf) 

fig.update_layout(template='plotly_white')

iplot(fig) 
```

```python
model1 = Prophet(yearly_seasonality = False, weekly_seasonality = False, daily_seasonality = True)
```

```python
# Dataframe must have columns "Date" and "Values" with the dates and values  for prophet prediction
Cured.columns = ['ds', 'y']
model1.fit(Cured)

future_1 = model1.make_future_dataframe(periods=60) # helper function to extend the dataframe for specified days
future_1.tail()
```

```python
# yhat represents the prediction, while yhat_lower and yhat_upper represent the lower and upper bound
forecast_cured = model1.predict(future_1)
forecast_cured
```

```python
fig = plot_plotly(model1, forecast_cured) 

fig.update_layout(template='plotly_white')

iplot(fig) 
```
