```python
import pandas as pd
import numpy as np
import datetime as dt
from fbprophet import Prophet
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import matplotlib.ticker as ticker
import logging
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline 
df = pd.read_csv('https://raw.githubusercontent.com/datasets/covid-19/master/data/countries-aggregated.csv', parse_dates=['Date'])
countries = ['US']
df = df[df['Country'].isin(countries)]
df['Cases'] = df[['Confirmed', 'Recovered', 'Deaths']].sum(axis=1)
```

```python
z = str("USA adopts social distancing")
z2 = str("All time high")
z3 = str("One million cases")
z4 = str("New Strain Confirmed in Country")
z5 = str("CDC Relaxes Mask Mandate")
z6 = str("USA Begins Vaccine Rollout")
```

```python
df = df.pivot(index='Date', columns='Country', values='Cases')
countries = list(df.columns)
covid = df.reset_index('Date')
covid.set_index(['Date'], inplace=True)
covid.columns = countries


#census.gov
populations = {'US':332452320}
percapita = covid.copy()
for country in list(percapita.columns):
    percapita[country] = percapita[country]/populations[country]*100000
```

```python
colors = {'US':'b'}
plt.style.use('fivethirtyeight')
plot = covid.plot(figsize=(18,12), color=list(colors.values()), linewidth=3, legend=True)
plt.title("COVID-19 Cases per capita")
plot.axvline(dt.datetime(2020, 3, 20),color='tab:green',linewidth=3,label=z).set_linestyle('dotted')
plot.axvline(dt.datetime(2021, 6, 7),color='tab:orange',linewidth=3,label=z2).set_linestyle('dotted')
plot.axvline(dt.datetime(2020, 4, 28),color='tab:pink',linewidth=3,label=z3).set_linestyle('dotted')
plot.axvline(dt.datetime(2020, 10, 29),color='tab:olive',linewidth=3,label=z4).set_linestyle('dotted')
plot.axvline(dt.datetime(2020, 5, 13),color='tab:purple',linewidth=3,label=z5).set_linestyle('dotted')
plot.axvline(dt.datetime(2020, 12, 14),color='tab:gray',linewidth=3,label=z6).set_linestyle('dotted')
plt.legend(loc= 'upper center')
plot.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
plot.grid(color='#FFFFFF')
plot.set_xlabel('Date')
plot.set_ylabel('# of Cases')
for country in list(colors.keys()):
    plot.text(x = covid.index[-1], y = covid[country].max(), color = colors[country], s = country, weight = 'bold')
```

```python
percapitaplot = percapita.plot(figsize=(20,10), color=list(colors.values()), linewidth=5, legend=True)
percapitaplot.grid(color='#FFFFFF')
percapitaplot.set_xlabel('Date')
percapitaplot.set_ylabel('# of Cases per 100,000 People')
plt.title("COVID-19 Cases per 100k People")
plt.legend(loc= 'upper center')
percapitaplot.axvline(dt.datetime(2020, 3, 20),color='tab:green',linewidth=3,label=z).set_linestyle('dotted')
percapitaplot.axvline(dt.datetime(2021, 6, 7),color='tab:orange',linewidth=3,label=z2).set_linestyle('dotted')
percapitaplot.axvline(dt.datetime(2020, 4, 28),color='tab:pink',linewidth=3,label=z3).set_linestyle('dotted')
percapitaplot.axvline(dt.datetime(2020, 10, 29),color='tab:olive',linewidth=3,label=z4).set_linestyle('dotted')
percapitaplot.axvline(dt.datetime(2020, 5, 13),color='tab:purple',linewidth=3,label=z5).set_linestyle('dotted')
percapitaplot.axvline(dt.datetime(2020, 11, 8),color='tab:gray',linewidth=3,label=z5).set_linestyle('dotted')
for country in list(colors.keys()):
    percapitaplot.text(x = percapita.index[-1], y = percapita[country].max(), color = colors[country], s = country, weight = 'bold')

```

```python
from fbprophet import Prophet
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline 
plt.rcParams['figure.figsize']=(20,10)
plt.style.use('ggplot')
```

```python
df.to_csv(r'US.csv', index = True, header=True)
df.head()
```

```python

df1 = pd.read_csv('US.csv', index_col='Date', parse_dates=True)

df1 = df1.reset_index().rename(columns={'Date':'ds', 'US':'y'})
#df2 = df1[df1.index % 7 == 1]
#df2 = df2.loc[(df2["ds"] > "2020-12-29")]
#df2['y_orig'] = df2['y'] 
df2['y'] = np.log(df1['y'])
#lets take a look at our data quickly
df2.set_index('ds').y.plot()
df2.head()
```

```python
model = Prophet(yearly_seasonality = False, weekly_seasonality = False, daily_seasonality = True, changepoint_range = .9)
model.fit(df2);
future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
model.plot(forecast)
```

```python
deltas = model.params['delta'].mean(0)
fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111)
ax.bar(range(len(deltas)), deltas)
ax.grid(True, which='major', c='gray', ls='-', lw=1, alpha=0.2)
ax.set_ylabel('Amount change')
ax.set_xlabel('Changepoint')
fig.tight_layout()
```
